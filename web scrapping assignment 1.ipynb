```python
#importing the libraries 
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
url="http://quotes.toscrape.com/page/1/"
```

```python
response=requests.get(url)
```

```python
response.content
```

```python
soup= BeautifulSoup(response.content,'html.parser')
```

```python
#print(soup.prettify())
```

```python
quote_containers = soup.find_all("div", class_="quote")

# Extract data from each quote container and store in a dictionary
data = {"author": [], "text": [], "tags": []}
for quote in quote_containers:
    data["author"].append(quote.find("small", class_="author").text)
    data["text"].append(quote.find("span", class_="text").text)
    tags = quote.find("div", class_="tags").find_all("a")
    data["tags"].append([tag.text for tag in tags])

# Convert the dictionary into a pandas dataframe
df = pd.DataFrame(data)
print(df)
```

```python
## getting the data of all 10 pages 
for i in range (1,10):
    url="http://quotes.toscrape.com/page/1/"
    response=requests.get(url)
    response.content
    soup= BeautifulSoup(response.content,'html.parser')
    quote_containers = soup.find_all("div", class_="quote")
    
    for quote in quote_containers:
        data["author"].append(quote.find("small", class_="author").text)
        data["text"].append(quote.find("span", class_="text").text)
        tags = quote.find("div", class_="tags").find_all("a")
        data["tags"].append([tag.text for tag in tags])
        
    df = pd.DataFrame(data)
    print(df)
```

```python
df.to_csv('data.csv', index=False)
```
